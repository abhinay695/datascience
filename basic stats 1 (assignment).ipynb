# 1.Mean:



 #The mean is the average of a set of numbers. It's calculated by adding up all the values in the set and dividing by the total number of values. For example, if you have a set of numbers {2, 4, 6, 8, 10}, the mean is (2 + 4 + 6 + 8 + 10) / 5 = 6. It gives you a sense of the central tendency of the data.


# 2.Median:
 

The median is the middle value of a dataset when it's ordered from least to greatest. If there's an even number of data points, it's the average of the two middle values. For example, in the set {3, 6, 7, 8, 8, 10, 14}, the median is 8. It's useful because it's not affected by extreme values (outliers) as much as the mean.

# 3.Mode:


The mode is the value that appears most frequently in a dataset. A dataset can have no mode (if no value is repeated), one mode (if one value is repeated more than any other), or multiple modes (if two or more values are repeated with the same frequency). For example, in the set {3, 5, 5, 7, 8, 8, 8, 10}, the mode is 8.

# 4.Standard Deviation:


The standard deviation measures the amount of variation or dispersion in a set of values. It quantifies how much the values in a dataset differ from the mean value. A low standard deviation indicates that the values tend to be close to the mean, while a high standard deviation indicates that the values are spread out over a wider range. It's calculated by taking the square root of the average of the squared differences between each value and the mean.

# boxplots

```python
import seaborn as sns
```

```python
numerical_columns = df.select_dtypes(include='number').columns
numerical_columns
plt.figure(figsize=(12,8))
sns.boxplot(data=df[numerical_columns])
plt.title("box plots for numercal columns")
plt.xticks(rotation=45)
plt.show()
```

```python
for column in numerical_columns:
    quartiles=df[column].quantile([0.25,0.75])
    iqr=quartiles[0.75]-quartiles[0.25]
    lower_bond=quartiles[0.25]-1.5*iqr
    upper_bond=quartiles[0.75]+1.5*iqr
    outliers=df[(df[column] < lower_bond) | (df[column] > upper_bond)]
    print(column)
    print(iqr)
    print(lower_bond)
    print(upper_bond)
    print(len(outliers))
    print(outliers)
    print("\n")
```

```python
categorical_columns=df.select_dtypes(include=["object"]).columns
categorical_columns
```

```python
for col in categorical_columns:
    plt.figure(figsize=(10,6))
    df[col].value_counts().plot(kind="bar",color="green")
    plt.title("bar charts for categorical_columns")
    plt.xlabel(col)
    plt.ylabel("frequency")
    plt.xticks(rotation=45)
    plt.show()
    print(f"analysis for column '{col}':")
    print(df[col].value_counts())
    print("\n")
```

# Standardization of Numerical Variables

# Standardization:
Standardization, or z-score normalization, is a method used to rescale data so that it has a mean of 0 and a standard deviation of 1. It allows for easier comparison of variables with different scales and distributions, making them more suitable for analysis and modelling.


```python
for column in numerical_columns:
    mean=df[column].mean()
    std=df[column].std()
    df[column+'_standardized']=(df[column]-mean)/std
    print(df)
```

# Conversion of Categorical Data into Dummy Variables

# Discuss the need for converting categorical data into dummy variables (one-hot encoding)................

Converting categorical data into dummy variables through one-hot encoding is necessary for machine learning algorithms because these algorithms typically work with numerical data. One-hot encoding transforms categorical variables into a format that algorithms can understand, allowing them to effectively interpret and process categorical information. This conversion ensures that all features contribute equally to the model, avoiding any inherent ordinality or magnitude assumptions associated with categorical variables.


```python
categorical_columns
```

```python
dummy_df=pd.get_dummies(df[categorical_columns],drop_first=True)
```

```python
preprosseed_df=pd.concat([df.drop(columns=categorical_columns),dummy_df],axis=1)
```
