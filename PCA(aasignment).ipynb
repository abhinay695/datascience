```python
#count the outliers in all varibles.

def count_outliers(df):
  outliers = {}
  for column in df:
    IQR = df[column].quantile(0.75) - df[column].quantile(0.25)
    lower_bound = df[column].quantile(0.25) - 1.5 * IQR
    upper_bound = df[column].quantile(0.75) + 1.5 * IQR
    outliers[column] = df[~((df[column] >= lower_bound) & (df[column] <= upper_bound))].shape[0]
  return outliers

outliers = count_outliers(df)
for column, count in outliers.items():
  print(f"Number of outliers in {column}: {count}")
```

```python
# replace the outliers with mean.

for column in df:
  IQR = df[column].quantile(0.75) - df[column].quantile(0.25)
  lower_bound = df[column].quantile(0.25) - 1.5 * IQR
  upper_bound = df[column].quantile(0.75) + 1.5 * IQR
  df[column] = df[column].apply(lambda x: df[column].mean() if x < lower_bound or x > upper_bound else x)
```

```python
#Investigate correlations between features to understand relationships within the data.

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Display the correlation matrix
print(correlation_matrix)

# Visualize the correlation matrix using a heatmap
import seaborn as sns
sns.heatmap(correlation_matrix, annot=True)
plt.show()

# Investigate specific correlations
high_correlation_features = []
for i in range(len(correlation_matrix.columns)):
  for j in range(i + 1, len(correlation_matrix.columns)):
    if abs(correlation_matrix.iloc[i, j]) > 0.7:
      high_correlation_features.append((correlation_matrix.columns[i], correlation_matrix.columns[j]))

print("Features with high correlation:")
for feature1, feature2 in high_correlation_features:
  print(f"{feature1} and {feature2}")
```

# Dimensionality Reduction with PCA

```python
#Standardize the features to ensure they have a mean of 0 and a standard deviation of Implement PCA to reduce the dimensionality of the dataset.

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Standardize the features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
df_scaled=pd.DataFrame(df_scaled)
col_names=["Type","Alcahol","Malic","Ash","Alcalinity","Magnesium","Phenols","F","N","P","C","H","D","P"]
df_scaled.columns=col_names
df_scaled
```

```python
# the optimal number of principal components using techniques like scree plot or cumulative explained variance.

# Scree plot
import numpy as np
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(df_scaled)
plt.plot(range(1, df_scaled.shape[1] + 1), pca.explained_variance_ratio_, marker='o')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.show()

# Cumulative explained variance
cumulative_explained_variance = np.cumsum(pca.explained_variance_ratio_)
plt.plot(range(1, df_scaled.shape[1] + 1), cumulative_explained_variance, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance')
plt.show()

# Choose the optimal number of principal components
optimal_components = 0
for i in range(1, df_scaled.shape[1] + 1):
  if cumulative_explained_variance[i] > 0.9:
    optimal_components = i
    break

print(f"Optimal number of principal components: {optimal_components}")
```

# Clustering with PCA Data

```python
X=pc[["Type","Alcahol"]]
X.shape
```

```python
# k-means clustering
#for k=3
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3,n_init=20)
Y = kmeans.fit_predict(X)

# Compute silhouette score
from sklearn.metrics import silhouette_score
silhouettescore = silhouette_score(X, Y)
silhouettescore
```

```python
#for loop for (2 to 10) k values
ss_score = []
for i in range(2,11,1):
    kmeans = KMeans(n_clusters=i,n_init=20)
    Y = kmeans.fit_predict(X)
    ss_score.append(silhouette_score(X, Y))

print(ss_score)

import matplotlib.pyplot as plt
plt.scatter(range(2,11,1),ss_score)
plt.plot(range(2,11,1),ss_score,color='red')
plt.xlabel("k vlaues")
plt.ylabel("s scores")
plt.show()
```


Practical Implications of PCA:

1. Dimensionality Reduction:
    
PCA can significantly reduce the computational complexity of data analysis tasks by reducing the number of features.
This is especially beneficial for large datasets with a high number of features, as it can make data analysis more feasible and efficient.

2. Improved Data Visualization:
    
PCA can be used to project high-dimensional data onto a lower-dimensional space for visualization purposes.
This can help to gain insights into the structure and distribution of the data, identify any potential outliers or misclassified data points, and explore relationships between different features.

3. Feature Selection:
                                                                              
PCA can be used to identify the most significant features in a dataset.
This can be useful for feature selection in various machine learning tasks, such as classification, regression, and clustering.
By focusing on the most informative features, PCA can help to improve the performance and interpretability of machine learning models.

4. Noise Reduction:
                                                                              
PCA can help to reduce the impact of noise and outliers on data analysis tasks.
By focusing on the most significant principal components, PCA can filter out irrelevant information and improve the quality of the analysis.

Practical Implications of Clustering:

1. Customer Segmentation:
                                                                              
Clustering can be used to segment customers into different groups based on their behavior, preferences, or other characteristics.
This information can be used to tailor marketing campaigns, product offerings, and customer service strategies to each segment.

2. Market Basket Analysis:
                                                                              
Clustering can be used to identify groups of items that are frequently purchased together.
This information can be used to improve product placement, design promotional campaigns, and develop personalized recommendations.

3. Fraud Detection:
                                                                              
Clustering can be used to identify unusual or fraudulent transactions based on their similarity to known fraudulent patterns.
This information can be used to flag suspicious transactions for further investigation.

4. Image Recognition:
                                    
Clustering can be used to group similar images together.
This information can be used to improve image retrieval systems, object detection algorithms, and facial recognition systems.

5. Anomaly Detection:
                                    
Clustering can be used to identify data points that are significantly different from the rest of the data.
This information can be used to detect anomalies, outliers, or errors in the data.

Overall, PCA and clustering are powerful tools that can be used to solve a wide range of data analysis problems.
PCA can be used to reduce dimensionality, improve data visualization, select features, and reduce noise.
Clustering can be used to segment customers, identify market basket patterns, detect fraud, recognize images, and detect anomalies.

By leveraging the capabilities of PCA and clustering, data analysts can gain valuable insights from their data and make informed decisions.
